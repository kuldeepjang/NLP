```python
import nltk
import numpy as np
import pandas as pd
```

```python
training_data=pd.read_csv('0000000000002747_training_twitter_x_y_train.csv')
testing_data=pd.read_csv('0000000000002747_test_twitter_x_test.csv')
```

```python
tweets=training_data['text']
tweets_test=testing_data['text']
```

```python
tweet_text=[]
for i in tweets:
    tweet_text.append(i)
```

```python
tweet_text_test=[]
for i in tweets_test:
    tweet_text_test.append(i)
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag): #creating simple tags to pass into the lemmatizer
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag
```

```python
output_words=[]
for text in tweet_text:
    words=text.split()
    clean_words=[]
    for i in words:
        if i.lower() not in stops:
            pos=pos_tag([i])
            clean_words.append(lemmatizer.lemmatize(i,get_simple_pos(pos[0][1])))
    output_words.append(clean_words)
```

```python
output_words_test=[]
for text in tweet_text_test:
    words=text.split()
    clean_words_test=[]
    for i in words:
        if i.lower() not in stops:
            pos=pos_tag([i])
            clean_words_test.append(lemmatizer.lemmatize(i,get_simple_pos(pos[0][1])))
    output_words_test.append(clean_words_test)
```

```python
y_train=training_data['airline_sentiment']
```

```python
x_train=[]
for i in range(len(output_words)):  
    joined=" ".join(output_words[i])
    x_train.append(joined)
```

```python
x_test=[]
for i in range(len(output_words_test)):  
    joined=" ".join(output_words_test[i])
    x_test.append(joined)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec=CountVectorizer(max_features=3000)
x_train_features=count_vec.fit_transform(x_train)
```

```python
x_train_features.todense()
```

```python
x_test_features=count_vec.transform(x_test)
```

```python
from sklearn.model_selection import cross_val_score,KFold,GridSearchCV
from sklearn .svm import SVC
from sklearn import svm
```

```python
clf=svm.SVC()
grid={'C':[1e2,1e4,1e1,5e3,1e3,5e2],'gamma':[1e-3,1e-2,5e-3,5e-2,1e-4,1e-5]}
abc=GridSearchCV(clf,grid, n_jobs=-1)
abc.fit(x_train_features,y_train)
```

```python
y_pred=abc.predict(x_test_features)
```
